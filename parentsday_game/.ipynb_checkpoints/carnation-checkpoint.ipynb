```python
import requests
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 웹에서 이미지 다운로드
image_url = 'https://blog.kakaocdn.net/dn/b0PTFe/btrz61iIbNa/4xFPAwjUrxA6htrLMYT9OK/img.png'  # 이미지를 가져올 URL을 입력하세요
response = requests.get(image_url)

# 이미지를 메모리에서 열기
img = mpimg.imread(BytesIO(response.content), format='jpg')

# 이미지 표시
plt.imshow(img)
plt.axis('off')  # 축을 숨기기
plt.show()
```
